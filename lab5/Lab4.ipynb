```python
def gradient_descent(start, learning_rate, iterations, gradient):
    x = start
    history  = np.array([x])
    for _ in range(iterations):
        grad = gradient(x)
        x -= learning_rate * grad
        history = np.append(history, x)
    return history
```

```python
def stochastic_gradient_descent(start, learning_rate, iterations, gradient, noise_type, sigma=0.01):
    x = start
    history  = np.array([x])
    for _ in range(iterations):
        grad = gradient(x)
        if noise_type=="normal":
            noise = np.random.normal(0,sigma)
        elif noise_type=="uniform":
            noise = np.random.uniform(-sigma, sigma)
        else:
            noise =0
        x -= learning_rate * grad + noise
        history = np.append(history, x)
    return history
```

```python
def plot(f, points, labels, bounds):
    plt.figure(figsize =(18,16))
    x = np.linspace(bounds[0],bounds[1],1000)
    plt.plot(x,f(x))
    for point, label in zip(points, labels):
        plt.plot(point, f(point), label = label, marker = ".")
    plt.xlim(bounds[0],bounds[1])
    #plt.ylim(f(bounds[0]),f(bounds[1]))
    plt.legend()
    plt.show()
```

```python
def gradient_descent2D(start, learning_rate, iterations, gradient):
    x, y = start
    history  = np.array([(start)])
    for _ in range(iterations):
        grad_x = gradient[0](x,y)
        grad_y = gradient[1](x,y)
        x -= learning_rate * grad_x
        y -= learning_rate * grad_y
        history = np.append(history, (x,y))
    return history.reshape(-1,2)
```

```python
def stochastic_gradient_descent2D(start, learning_rate, iterations, gradient, noise_type, sigma=0.05):
    x, y = start
    history  = np.array([(start)])
    for _ in range(iterations):
        if noise_type=="normal":
            noise_x = np.random.normal(0,sigma)
            noise_y = np.random.normal(0,sigma)
        elif noise_type=="uniform":
            noise_x = np.random.uniform(-sigma, sigma)
            noise_y = np.random.uniform(-sigma, sigma)
        else:
            noise =0
        grad_x = gradient[0](x,y)
        grad_y = gradient[1](x,y)
        x -= learning_rate * grad_x + noise_x
        y -= learning_rate * grad_y + noise_y
        history = np.append(history, (x,y))
    return history.reshape(-1,2)
```

```python
def plot2D(history_list, labels, function, xlim, ylim):
    h=0.1
    x_min, x_max = xlim.min(), xlim.max()
    y_min, y_max = ylim.min(), ylim.max()
    X = np.arange(x_min, x_max, h)
    Y = np.arange(y_min, y_max, h)
    xx, yy = np.meshgrid(X, Y)
    Z = function(xx,yy)
    plt.contour(xx,yy,Z,levels =1000)
    for point, label in zip(history_list, labels):
        plt.plot(point[:,0],point[:,1], label = label, marker = ".",alpha=0.5)
    plt.legend()
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.show()

    ax = plt.figure().add_subplot(projection='3d')

    
    # Plot the 3D surface
    ax.plot_surface(xx,yy,Z, edgecolor='royalblue', lw=0.05, rstride=1, cstride=8,
                    alpha=0.3)
    for point, label in zip(history_list, labels):
        ax.plot(point[:,0],point[:,1], function(point[:,0],point[:,1]), label = label, marker = ".",alpha=0.5)
    plt.legend()
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.show()
    plt.show()
```

### Momentum

```python
def momentum_gd_1d(start, learning_rate, iterations, gradient, momentum=0.9):
    x = start
    history = np.array([start])
    velocity = 0
    for _ in range(iterations):
        grad = gradient(x)
        velocity = momentum * velocity - learning_rate * grad
        x += velocity
        history = np.append(history, x)
    return history
```

### RMSProp

```python
def RMSProp(start, learning_rate, iterations, gradient, beta=0.9, epsilon=1e-6):
    x = start
    history = np.array([start])
    grad_squared = 0
    for _ in range(iterations):
        grad = gradient(x)
        grad_squared = beta * grad_squared + (1 - beta) * np.square(grad)
        x -= learning_rate * grad / (np.sqrt(grad_squared + epsilon))
        history = np.append(history, x)
    return history
```

## ADAM

```python
def ADAM(start, learning_rate, iterations, gradient, beta1 = 0.9, beta2 = 0.9999, epsilon = 1e-6):
    x = start
    history = np.array([start])
    m = 0
    v = 0
    for t in range(1, iterations + 1):
        grad = gradient(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * np.square(grad)
        x -= learning_rate * m / (np.sqrt(v) + epsilon)
        history = np.append(history, x)
    return history
```

```python
def comp_gd_sgd_adam_rmsprop_momentum_plots(start, f, iterations, gradient, bounds):
    lr = 0.001
    history_gd = gradient_descent(start, lr, iterations, gradient) 
    history_sgd_n = stochastic_gradient_descent(start, lr, iterations, gradient,"normal",sigma=0.01) 
    history_sgd_u = stochastic_gradient_descent(start, lr, iterations, gradient,"uniform",sigma=0.01) 
    history_momentum = momentum_gd_1d(start, lr, iterations, gradient)
    history_rmsprop = RMSProp(start, lr, iterations, gradient)
    history_adam = ADAM(start, lr, iterations, gradient)
    plot(f,[history_gd,history_sgd_n,history_sgd_u,
            history_momentum,
            history_rmsprop,
            history_adam], 
         ["Gradient Descent","Stochastic Gradient Descent (Normal noise)","S Gradient Descent (Uniform noise)",
          "Momentum GD", "RMSProp", "ADAM"], bounds)
    x = np.arange(1, iterations+2)
    plt.figure(figsize =(15,12))
    plt.plot(x, history_gd, label = "GD")
    plt.plot(x, history_sgd_n, label = "SGD N")
    plt.plot(x, history_sgd_u, label = "SGD U")
    plt.plot(x, history_momentum, label = "Momentum GD")
    plt.plot(x, history_rmsprop, label = "RMSProp")
    plt.plot(x, history_adam, label = "ADAM")
    plt.xlabel("Iteration")
    plt.ylabel("x val")
    plt.legend()
```

```python
import matplotlib.pyplot as plt

lr_vals = np.linspace(0.0001, 0.5, 60)
beta1_vals = np.linspace(0.8, 0.999999, 60)
beta2_vals = np.linspace(0.8, 0.999999, 60)

results = []

for lr in lr_vals[::3]:
    for beta1 in beta1_vals[::3]:
        for beta2 in beta2_vals[::3]:
            history_adam = ADAM(10, lr, 1000, f3_prim, beta1=beta1, beta2=beta2, epsilon=1e-6)
            
            
            results.append((lr, beta1, beta2, history_adam[-1]))

plt.figure(figsize=(10, 8))
plt.plot(f3(np.array(history_adam)), label="Function Value")
plt.title(f"Function Value for ADAM (lr: {lr:.4f}, 1: {beta1:.4f}, 2: {beta2:.4f})")
plt.xlabel("Iterations")
plt.ylabel("f3(x)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

if results:
    best = min(results, key=lambda x: abs(x[3]))
    print(f"Best parameters found: lr={best[0]:.4f}, beta1={best[1]:.4f}, beta2={best[2]:.4f}, final_value={best[3]:.4f}")
```
