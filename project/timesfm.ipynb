```python
import pandas as pd
import matplotlib.pyplot as plt
import timesfm
import numpy as np
from tqdm.notebook import tqdm
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

DATA_FILE_PATH = '/kaggle/input/df3333/df3.csv'
INITIAL_FORECAST_START_DATE = pd.to_datetime("2022-02-24")
SIMULATION_END_DATE = pd.to_datetime("2025-04-04")
ROLLING_HORIZON_LEN = 30
CONTEXT_LEN = 480 
STEP_SIZES_TO_TEST = [1,5,10,15,20] 
TIMESFM_MODEL_REPO = "google/timesfm-2.0-500m-pytorch"
BACKEND = "gpu" 

print("Loading and preparing data...")
df_original = pd.read_csv(DATA_FILE_PATH)
df_original['Date'] = pd.to_datetime(df_original['Date'])
df_original.sort_values('Date', inplace=True)

df_original.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)
print(f"Data loaded. Date range: {df_original['ds'].min()} to {df_original['ds'].max()}")
print(f"Number of rows: {len(df_original)}")

print(f"Initializing TimesFM model ({TIMESFM_MODEL_REPO})...")
tfm = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend=BACKEND,
          per_core_batch_size=32,
          horizon_len=ROLLING_HORIZON_LEN,
          num_layers=50,
          use_positional_embedding=False,
          context_len=CONTEXT_LEN,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id=TIMESFM_MODEL_REPO),
)
assert tfm.hparams.context_len == CONTEXT_LEN, "Context Length mismatch!"
print("TimesFM model initialized.")

def run_rolling_forecast_simulation(
    step_size_days,
    tfm_model,
    full_data_df, 
    sim_start_date,
    sim_end_date,
    context_len,
    horizon_len 
):
    print(f"\n--- Starting simulation for step_size = {step_size_days} days ---")

    initial_history_end_date = sim_start_date - pd.Timedelta(days=1)
    df_initial_history = full_data_df[full_data_df['ds'] <= initial_history_end_date].copy()

    if len(df_initial_history) < context_len:
        print(f"CRITICAL ERROR: Insufficient initial history ({len(df_initial_history)} days) for context_len={context_len}. Aborting simulation for step_size={step_size_days}.")
        return None, None, None 

    all_step_forecasts = []
    current_history_end_date = initial_history_end_date
    date_range = pd.date_range(start=sim_start_date, end=sim_end_date, freq=f'{step_size_days}D')

    print(f"Simulation from {sim_start_date.strftime('%Y-%m-%d')} to {sim_end_date.strftime('%Y-%m-%d')} (step: {step_size_days} days)")
    for current_forecast_start_date in tqdm(date_range, desc=f"Step={step_size_days}d"):
        history_for_step = full_data_df[
            (full_data_df['ds'] <= current_history_end_date)
        ].tail(context_len).copy()

        if len(history_for_step) < context_len:
            print(f"WARNING: In step for {current_forecast_start_date}, available history ({len(history_for_step)}) is shorter than context_len ({context_len}). Probably end of data. Stopping.")
            break

        input_df_step = pd.DataFrame({
            "unique_id": ["TSLA"] * len(history_for_step),
            "ds": history_for_step["ds"],
            "y": history_for_step["y"].values.flatten()
        })

        try:
            forecast_df_step = tfm_model.forecast_on_df(
                inputs=input_df_step,
                freq="D",
                value_name="y",
                num_jobs=-1
            )
            if forecast_df_step is None or forecast_df_step.empty:
                 print(f"WARNING: Forecast for step ending on {current_history_end_date} is empty. Skipping step.")
                 current_history_end_date += pd.Timedelta(days=step_size_days)
                 continue

        except Exception as e:
            print(f"ERROR during forecasting for history ending on {current_history_end_date}: {e}")
            current_history_end_date += pd.Timedelta(days=step_size_days)
            continue

        next_steps_forecast = forecast_df_step[
            forecast_df_step['ds'] >= current_forecast_start_date
        ].head(step_size_days).copy()

        if not next_steps_forecast.empty:
             next_steps_forecast['generated_after'] = current_history_end_date
             all_step_forecasts.append(next_steps_forecast)

        current_history_end_date += pd.Timedelta(days=step_size_days)

        if current_history_end_date > full_data_df['ds'].max():
             print(f"Reached the end of available historical data ({full_data_df['ds'].max()}). Stopping loop for step_size={step_size_days}.")
             break

    print(f"Finished forecasting loop for step_size = {step_size_days} days.")

    if not all_step_forecasts:
        print(f"No forecasts were generated for step_size = {step_size_days}.")
        return None, None, None

    final_rolling_forecast_df = pd.concat(all_step_forecasts, ignore_index=True)
    final_rolling_forecast_df.sort_values(by=['generated_after', 'ds'], inplace=True)
    final_rolling_forecast_df.drop_duplicates(subset=['ds'], keep='first', inplace=True)

    comparison_df = pd.merge(
        final_rolling_forecast_df[['ds', 'timesfm']],
        full_data_df[['ds', 'y']],
        on='ds',
        how='inner'
    )

    if comparison_df.empty:
        print(f"No common dates found to calculate metrics for step_size = {step_size_days}.")
        mae, rmse = np.nan, np.nan
    else:
        y_pred = comparison_df["timesfm"]
        y_true = comparison_df["y"]
        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        print(f"Metrics for step_size = {step_size_days} (on {len(comparison_df)} points): MAE={mae:.2f}, RMSE={rmse:.2f}")

    return final_rolling_forecast_df, mae, rmse

def plot_rolling_comparison(forecast_df, full_data_df, sim_start_date, step_size):
    plt.figure(figsize=(18, 8))

    initial_history_end_date = sim_start_date - pd.Timedelta(days=1)
    df_initial_history = full_data_df[full_data_df['ds'] <= initial_history_end_date]
    plt.plot(df_initial_history["ds"], df_initial_history["y"], label="Historical Price (before forecast)", color="black", linewidth=1.5)

    plt.plot(forecast_df["ds"], forecast_df["timesfm"], label=f"Rolling Median Forecast (step={step_size}d)", color="blue", linewidth=1.5)

    if 'timesfm-q-0.3' in forecast_df.columns and 'timesfm-q-0.7' in forecast_df.columns:
        plt.fill_between(
            forecast_df["ds"],
            forecast_df["timesfm-q-0.3"],
            forecast_df["timesfm-q-0.7"],
            color="blue",
            alpha=0.2,
            label="Rolling Forecast Interval (30-70%)"
        )
    else:
        print("Warning: No quantile columns in forecast results. Interval will not be drawn.")

    df_actual_compare = full_data_df[
        (full_data_df['ds'] >= forecast_df['ds'].min()) &
        (full_data_df['ds'] <= forecast_df['ds'].max())
    ]
    plt.plot(df_actual_compare["ds"], df_actual_compare["y"], label="Actual Price (during forecast period)", color="red", linestyle="--", linewidth=1.5)

    plot_start_date_vis = initial_history_end_date - pd.Timedelta(days=90)
    plot_end_date_vis = forecast_df["ds"].max()
    plt.xlim([plot_start_date_vis, plot_end_date_vis])

    plt.xlabel("Date")
    plt.ylabel("Stock Price")
    plt.title(f"Rolling Forecast (every {step_size} days) vs Reality (start: {sim_start_date.strftime('%Y-%m-%d')})")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

results_list = []

for step in STEP_SIZES_TO_TEST:
    forecast_result_df, result_mae, result_rmse = run_rolling_forecast_simulation(
        step_size_days=step,
        tfm_model=tfm,
        full_data_df=df_original,
        sim_start_date=INITIAL_FORECAST_START_DATE,
        sim_end_date=SIMULATION_END_DATE,
        context_len=CONTEXT_LEN,
        horizon_len=ROLLING_HORIZON_LEN
    )

    if forecast_result_df is not None and not forecast_result_df.empty:
        plot_rolling_comparison(
            forecast_df=forecast_result_df,
            full_data_df=df_original,
            sim_start_date=INITIAL_FORECAST_START_DATE,
            step_size=step
        )
        results_list.append({
            'step_size (days)': step,
            'MAE': result_mae,
            'RMSE': result_rmse,
            'forecast_end_date': forecast_result_df['ds'].max()
        })
    else:
         results_list.append({
            'step_size (days)': step,
            'MAE': np.nan,
            'RMSE': np.nan,
            'forecast_end_date': pd.NaT
        })

results_df = pd.DataFrame(results_list)

print("\n--- Final Metrics Results ---")
print(results_df)
```
