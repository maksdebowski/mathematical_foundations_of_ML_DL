```python
import pandas as pd
import matplotlib.pyplot as plt
import timesfm
import numpy as np
from tqdm.notebook import tqdm
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

DATA_FILE_PATH = '/kaggle/input/df3333/df3.csv'
INITIAL_FORECAST_START_DATE = pd.to_datetime("2022-02-24")
SIMULATION_END_DATE = pd.to_datetime("2025-04-04")
ROLLING_HORIZON_LEN = 30
CONTEXT_LEN = 128     
STEP_SIZES_TO_TEST = [1, 5, 10, 15, 20] 
TIMESFM_MODEL_REPO = "google/timesfm-2.0-500m-pytorch"
BACKEND = "gpu"
TARGET_VARIABLE = 'Realized_Volatility_Daily' 
UNIQUE_ID = "RHM" 

def qlike_loss(y_true, y_pred, epsilon=1e-8):
    """Oblicza stratę QLIKE."""
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = (y_true > epsilon) & (y_pred > epsilon)
    if mask.sum() == 0:
        return np.nan
    y_true_m, y_pred_m = y_true[mask], y_pred[mask]
    ratio = y_true_m / y_pred_m
    loss = ratio - np.log(ratio) - 1
    return np.mean(loss)

print("Loading and preparing data...")
df_original = pd.read_csv(DATA_FILE_PATH)
df_original['Date'] = pd.to_datetime(df_original['Date'])
df_original.sort_values('Date', inplace=True)
df_original.rename(columns={'Date': 'ds'}, inplace=True)

print("Calculating Log Returns and Realized Volatility...")
df_original['Adj Close'] = df_original['Adj Close'].replace(0, np.nan).ffill() 
df_original['Log_Return'] = np.log(df_original['Adj Close'] / df_original['Adj Close'].shift(1))

window = 21 
annualization_factor = 260

df_original[TARGET_VARIABLE] = df_original['Log_Return'].rolling(window=window).std() * np.sqrt(annualization_factor) * 100

initial_rows = len(df_original)
df_original.dropna(subset=['Log_Return', TARGET_VARIABLE], inplace=True)
print(f"Data loaded. Date range: {df_original['ds'].min()} to {df_original['ds'].max()}")
print(f"Number of rows after calculating volatility and dropping NaNs: {len(df_original)} (removed {initial_rows - len(df_original)} rows)")

print(f"Initializing TimesFM model ({TIMESFM_MODEL_REPO})...")
tfm = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend=BACKEND,
          per_core_batch_size=32,
          horizon_len=ROLLING_HORIZON_LEN,
          num_layers=50,
          use_positional_embedding=False, 
          context_len=CONTEXT_LEN,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id=TIMESFM_MODEL_REPO),
)
assert tfm.hparams.context_len == CONTEXT_LEN, "Context Length mismatch!"
print("TimesFM model initialized.")

def run_rolling_forecast_simulation(
    step_size_days,
    tfm_model,
    full_data_df,
    target_col_name, 
    sim_start_date,
    sim_end_date,
    context_len,
    horizon_len,
    unique_id_val 
):
    print(f"\n--- Starting simulation for step_size = {step_size_days} days (Target: {target_col_name}) ---")

    initial_history_end_date = sim_start_date - pd.Timedelta(days=1)

    df_initial_history_check = full_data_df[full_data_df['ds'] <= initial_history_end_date]
    if len(df_initial_history_check) < context_len:
        print(f"CRITICAL ERROR: Insufficient initial history ({len(df_initial_history_check)} days available before {sim_start_date.strftime('%Y-%m-%d')}) for context_len={context_len}. Aborting simulation for step_size={step_size_days}.")
        return None, np.nan, np.nan, np.nan 

    all_step_forecasts = []
    current_history_end_date = initial_history_end_date
    available_dates_in_range = full_data_df[(full_data_df['ds'] >= sim_start_date) & (full_data_df['ds'] <= sim_end_date)]['ds']
    if available_dates_in_range.empty:
         print(f"WARNING: No available data points between {sim_start_date.strftime('%Y-%m-%d')} and {sim_end_date.strftime('%Y-%m-%d')}. Cannot run simulation.")
         return None, np.nan, np.nan, np.nan

    date_range = pd.date_range(start=available_dates_in_range.min(), end=available_dates_in_range.max(), freq=f'{step_size_days}D')
    actual_forecast_start_dates = []
    last_used_date = pd.Timestamp.min
    for potential_start_date in date_range:
        closest_date = available_dates_in_range[available_dates_in_range >= potential_start_date].min()
        if pd.notna(closest_date) and closest_date > last_used_date:
             actual_forecast_start_dates.append(closest_date)
             last_used_date = closest_date + pd.Timedelta(days=step_size_days - 1)


    if not actual_forecast_start_dates:
         print(f"WARNING: Could not determine valid forecast start dates in the range. Aborting.")
         return None, np.nan, np.nan, np.nan

    print(f"Effective simulation from {actual_forecast_start_dates[0].strftime('%Y-%m-%d')} to {actual_forecast_start_dates[-1].strftime('%Y-%m-%d')} using {len(actual_forecast_start_dates)} steps.")

    for current_forecast_start_date in tqdm(actual_forecast_start_dates, desc=f"Step={step_size_days}d"):
        current_history_end_date = current_forecast_start_date - pd.Timedelta(days=1)

        history_for_step = full_data_df[
            (full_data_df['ds'] <= current_history_end_date)
        ].tail(context_len).copy()

        if len(history_for_step) < context_len:
            print(f"WARNING: In step for {current_forecast_start_date.strftime('%Y-%m-%d')}, available history ({len(history_for_step)}) is shorter than context_len ({context_len}). Stopping.")
            break

        input_df_step = pd.DataFrame({
            "unique_id": [unique_id_val] * len(history_for_step),
            "ds": history_for_step["ds"],
            target_col_name: history_for_step[target_col_name].values.flatten() 
        })

        try:
            forecast_df_step = tfm_model.forecast_on_df(
                inputs=input_df_step,
                freq="D", 
                value_name=target_col_name,
                num_jobs=-1
            )
            if forecast_df_step is None or forecast_df_step.empty:
                 print(f"WARNING: Forecast generated on {current_history_end_date.strftime('%Y-%m-%d')} for period starting {current_forecast_start_date.strftime('%Y-%m-%d')} is empty. Skipping step.")
                 continue

        except Exception as e:
            print(f"ERROR during forecasting based on history ending {current_history_end_date.strftime('%Y-%m-%d')}: {e}")
            continue

        forecast_df_step = forecast_df_step.sort_values('ds')
        next_steps_forecast = forecast_df_step[
            forecast_df_step['ds'] >= current_forecast_start_date
        ].head(step_size_days).copy()

        if not next_steps_forecast.empty:
             next_steps_forecast['generated_after'] = current_history_end_date
             all_step_forecasts.append(next_steps_forecast)
 



    print(f"Finished forecasting loop for step_size = {step_size_days} days.")

    if not all_step_forecasts:
        print(f"No forecasts were generated for step_size = {step_size_days}.")
        return None, np.nan, np.nan, np.nan

    final_rolling_forecast_df = pd.concat(all_step_forecasts, ignore_index=True)

    final_rolling_forecast_df.sort_values(by=['ds', 'generated_after'], inplace=True)
    final_rolling_forecast_df.drop_duplicates(subset=['ds'], keep='first', inplace=True)
    final_rolling_forecast_df.sort_values(by='ds', inplace=True)

    comparison_df = pd.merge(
        final_rolling_forecast_df[['ds', 'timesfm']], 
        full_data_df[['ds', target_col_name]],      
        on='ds',
        how='inner'
    )

    if comparison_df.empty:
        print(f"No common dates found to calculate metrics for step_size = {step_size_days}.")
        mae, rmse, qlike = np.nan, np.nan, np.nan
    else:
        y_pred = comparison_df["timesfm"]
        y_true = comparison_df[target_col_name]

        valid_comparison = comparison_df.dropna()
        if valid_comparison.empty:
             print(f"No valid (non-NaN) comparison points found for step_size = {step_size_days}.")
             mae, rmse, qlike = np.nan, np.nan, np.nan
        else:
            y_pred_valid = valid_comparison["timesfm"]
            y_true_valid = valid_comparison[target_col_name]

            mae = mean_absolute_error(y_true_valid, y_pred_valid)
            rmse = np.sqrt(mean_squared_error(y_true_valid, y_pred_valid))
            qlike = qlike_loss(y_true_valid, y_pred_valid) 
            print(f"Metrics for step_size = {step_size_days} (on {len(valid_comparison)} points): MAE={mae:.4f}, RMSE={rmse:.4f}, QLIKE={qlike:.4f}")

    return final_rolling_forecast_df, mae, rmse, qlike

def plot_rolling_comparison(forecast_df, full_data_df, target_col_name, sim_start_date, step_size):
    plt.figure(figsize=(18, 8))

    initial_history_end_date = sim_start_date - pd.Timedelta(days=1)
    df_initial_history = full_data_df[full_data_df['ds'] <= initial_history_end_date]
    plt.plot(df_initial_history["ds"], df_initial_history[target_col_name], label=f"Historical {target_col_name} (before forecast)", color="black", linewidth=1.5)

    plt.plot(forecast_df["ds"], forecast_df["timesfm"], label=f"Rolling Median Forecast (step={step_size}d)", color="blue", linewidth=1.5)

    if 'timesfm-q-0.3' in forecast_df.columns and 'timesfm-q-0.7' in forecast_df.columns:
        plt.fill_between(
            forecast_df["ds"],
            forecast_df["timesfm-q-0.3"],
            forecast_df["timesfm-q-0.7"],
            color="blue",
            alpha=0.2,
            label="Rolling Forecast Interval (30-70%)"
        )

    df_actual_compare = full_data_df[
        (full_data_df['ds'] >= forecast_df['ds'].min()) &
        (full_data_df['ds'] <= forecast_df['ds'].max())
    ]
    plt.plot(df_actual_compare["ds"], df_actual_compare[target_col_name], label=f"Actual {target_col_name} (during forecast period)", color="red", linestyle="--", linewidth=1.5)

    plot_start_date_vis = initial_history_end_date - pd.Timedelta(days=90)
    plot_end_date_vis = forecast_df["ds"].max() + pd.Timedelta(days=10)
    plt.xlim([plot_start_date_vis, plot_end_date_vis])

    plt.xlabel("Date")
    plt.ylabel(f"{target_col_name}")
    plt.title(f"TimesFM Rolling Forecast (every {step_size} days) vs Reality for {target_col_name} (start: {sim_start_date.strftime('%Y-%m-%d')})")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

results_list = []

for step in STEP_SIZES_TO_TEST:
    forecast_result_df, result_mae, result_rmse, result_qlike = run_rolling_forecast_simulation(
        step_size_days=step,
        tfm_model=tfm,
        full_data_df=df_original,
        target_col_name=TARGET_VARIABLE, 
        sim_start_date=INITIAL_FORECAST_START_DATE,
        sim_end_date=SIMULATION_END_DATE,
        context_len=CONTEXT_LEN,
        horizon_len=ROLLING_HORIZON_LEN,
        unique_id_val=UNIQUE_ID 
    )

    if forecast_result_df is not None and not forecast_result_df.empty:
        plot_rolling_comparison(
            forecast_df=forecast_result_df,
            full_data_df=df_original,
            target_col_name=TARGET_VARIABLE, 
            sim_start_date=INITIAL_FORECAST_START_DATE,
            step_size=step
        )
        results_list.append({
            'step_size (days)': step,
            'MAE_vol': result_mae,
            'RMSE_vol': result_rmse,
            'QLIKE_vol': result_qlike, 
            'forecast_end_date': forecast_result_df['ds'].max()
        })
    else:
         results_list.append({
            'step_size (days)': step,
            'MAE_vol': np.nan,
            'RMSE_vol': np.nan,
            'QLIKE_vol': np.nan,
            'forecast_end_date': pd.NaT
        })

results_df = pd.DataFrame(results_list)

print(f"\n--- Final Metrics Results for {TARGET_VARIABLE} ---")
print(results_df)
```
