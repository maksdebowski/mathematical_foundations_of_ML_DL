```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class f:
    def f(x: float):
        return x*x

    def prim(x: float):
        return 2*x

def GD(start: float, f: f,n: int, learning_rate: float):
    results = np.zeros(n)
    results[0] = start
    for i in range(1, n):
        grad = f.prim(results[i-1])
        results[i] = results[i-1] - learning_rate * grad
    return results

def plot(start: float, f: f, iterations: int, learning_rate: float):
    x_vals = np.linspace(start, - start, 100)
    y_vals = f.f(x_vals)
    plt.plot(x_vals, y_vals)
    trajectory = GD(start, f, iterations, learning_rate)
    trajectory_y = f.f(trajectory)
    plt.plot(trajectory, trajectory_y, 'ro-')
    plt.grid(True)
    plt.show()

def plot_convergence(start:float, f: f, iterations, learning_rate):
    x_vals = GD(start, f, iterations, learning_rate)
    plt.plot(range(iterations), f.f(x_vals), 'ro-', markersize=3)
    plt.plot(range(iterations), x_vals, 'b')
    plt.xlabel('iterations')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()
```

```python
plot(-2, f, 10, 0.9)
plot_convergence(-2, f, 50, 0.9)
```

```python
class f:
    def f(x: float):
        return x**3 - 10*(x**2)

    def prim(x: float):
        return 3*(x**2) - 20*x


plot(1, f, 1000, 0.001)
plot_convergence(1, f, 1000, 0.001)
```

```python
class f:
    def f(x: float):
        return x*x

    def prim(x: float):
        return 2*x

def SGD(start: float, f: f,n: int, learning_rate: float, sigma: float, noise = "Normal" or "Uniform"):
    results = np.zeros(n)
    results[0] = start
    for i in range(1, n):
        grad = f.prim(results[i-1])
        if noise == 'Normal':
            noise = np.random.normal(-sigma, sigma)
        elif noise == 'Uniform':
            noise = np.random.uniform(-sigma, sigma)
        results[i] = results[i-1] - learning_rate * grad + noise
    return results

def plot(start: float, f: f, iterations: int, learning_rate: float, sigma: float, noise = "Normal" or "Uniform"):
    x_vals = np.linspace(start, - start, 100)
    y_vals = f.f(x_vals)
    plt.plot(x_vals, y_vals)
    trajectory = SGD(start, f, iterations, learning_rate, sigma, noise)
    trajectory_y = f.f(trajectory)
    plt.plot(trajectory, trajectory_y)
    plt.grid(True)
    plt.show()

def plot_convergence(start:float, f: f, iterations, learning_rate, sigma: float, noise = "Normal" or "Uniform"):
    x_vals = SGD(start, f, iterations, learning_rate, sigma, noise)
    plt.plot(range(iterations), f.f(x_vals), 'ro-', markersize=3)
    plt.plot(range(iterations), x_vals, 'b')
    plt.xlabel('iterations')
    plt.ylabel('f(x)')
    plt.title(f"{noise}")
    plt.grid(True)
    plt.show()
```

```python
plot(-2, f, 10, 0.9, 0.01, "Normal")
plot_convergence(-2, f, 10, 0.9, 0.01, "Normal")
```
